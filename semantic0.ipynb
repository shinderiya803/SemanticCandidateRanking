```python
from docx import Document # Read job description

doc = Document("/content/data/[PUB] India_runs_data_and_ai_challenge/India_runs_data_and_ai_challenge/job_description.docx")

job_description = ""

for para in doc.paragraphs:
    job_description += para.text + "\n"

print(job_description)
```

```python
# Load candidates Data sets
import json

candidates = []

with open("/content/data/[PUB] India_runs_data_and_ai_challenge/India_runs_data_and_ai_challenge/candidates.jsonl", "r", encoding="utf-8") as f:
    for line in f:
        candidates.append(json.loads(line))

print("Number of candidates:", len(candidates))
```

```python
#inspect the first Candidate
candidates[0]
```

```python
def candidate_to_text(candidate):

    profile = candidate['profile']

    headline = profile.get('headline', '')
    summary = profile.get('summary', '')
    title = profile.get('current_title', '')
    experience = str(profile.get('years_of_experience', ''))

    skills = " ".join(
        skill['name']
        for skill in candidate['skills']
    )

    career = " ".join(
        job['description']
        for job in candidate['career_history']
    )

    text = f"""
    {headline}

    {summary}

    Current Role:
    {title}

    Experience:
    {experience} years

    Skills:
    {skills}

    Career History:
    {career}
    """

    return text
```

```python
#Load Model
from sentence_transformers import SentenceTransformer

model = SentenceTransformer('all-MiniLM-L6-v2')
```

```python
# convert all candidates into text
candidate_texts = [candidate_to_text(candidate)for candidate in candidates]

print(candidate_texts[0])
```

```python
# Generate the job embedding
job_embedding = model.encode(job_description,normalize_embeddings=True)
```

```python
#define Relevant Keyword
good_keywords = [ "retrieval","ranking","recommendation", "embedding","sentence-transformers", "faiss", "milvus","qdrant","weaviate",
                 "pinecone","elasticsearch","opensearch","llm", "fine-tuning", "lora","peft","rag","vector", "search"]
```

```python
#create candidate document
def candidate_document(candidate):

    profile = candidate['profile']

    text = ""

    text += profile.get('headline', '') + " "
    text += profile.get('summary', '') + " "

    for job in candidate['career_history']:
        text += job.get('title', '') + " "
        text += job.get('description', '') + " "

    for skill in candidate['skills']:
        text += skill['name'] + " "

    return text.lower()
```

```python
#filter Candidate
filtered_indices = []

for i, candidate in enumerate(candidates):

    years = candidate['profile']['years_of_experience']

    if years < 4 or years > 10:
        continue

    doc = candidate_document(candidate)

    score = sum(
        keyword in doc
        for keyword in good_keywords
    )

    if score >= 2:
        filtered_indices.append(i)

print("Filtered candidates:", len(filtered_indices))
```

```python
#Create filter Candidate list
filtered_candidates = [
    candidates[i]
    for i in filtered_indices
]

print(len(filtered_candidates))
```

```python
# Create candidate Texts
filtered_candidate_texts = [
      candidate_to_text(candidate)
    for candidate in filtered_candidates
]

print(filtered_candidate_texts[0][:1000])
```

```python
# Generate Embedding
import numpy as np

filtered_embeddings = model.encode(
    filtered_candidate_texts,
    batch_size=512,
    normalize_embeddings=True,
    show_progress_bar=True
)

filtered_embeddings = np.array(filtered_embeddings)

print(filtered_embeddings.shape)
```

```python
# compute Semantic similarities
similarities = filtered_embeddings @ job_embedding
```

```python
#Get Top 500 Candidates
top_k = 500

top_indices = np.argsort(similarities)[::-1][:top_k]

print(top_indices[:10])
```

```python
# Create Top Candidate list
top_candidates = [
    filtered_candidates[i]
    for i in top_indices
]

print(len(top_candidates))
```

```python
#Skill Score
core_skills = ["embedding","retrieval","ranking","llm","fine-tuning","faiss","milvus",
               "qdrant","weaviate","pinecone","python",  "lora","peft", "search"]

def skill_score(candidate):

    skills = [
        skill['name'].lower()
        for skill in candidate['skills']
    ]

    matches = sum(
        keyword in " ".join(skills)
        for keyword in core_skills
    )

    return matches / len(core_skills)
```

```python
#Experience Score
def experience_score(candidate):

    years = candidate['profile']['years_of_experience']

    if 5 <= years <= 9:
        return 1.0

    elif 4 <= years < 5:
        return 0.8

    elif 9 < years <= 10:
        return 0.8

    else:
        return 0.4
```

```python
#redrobe Score
def redrob_score(candidate):

    signals = candidate['redrob_signals']

    score = 0

    if signals['open_to_work_flag']:
        score += 0.2

    score += (
        signals['recruiter_response_rate']
    ) * 0.2

    score += (
        signals['interview_completion_rate']
    ) * 0.2

    score += (
        signals['offer_acceptance_rate']
    ) * 0.2

    score += (
        signals['github_activity_score'] / 100
    ) * 0.2

    return score
```

```python
#Combine Everything
ranked_candidates = []

for idx in top_indices:

    candidate = filtered_candidates[idx]

    semantic = similarities[idx]

    skill = skill_score(candidate)

    exp = experience_score(candidate)

    redrob = redrob_score(candidate)

    final_score = (
        0.4 * semantic
        + 0.25 * skill
        + 0.15 * exp
        + 0.20 * redrob
    )

    ranked_candidates.append(
        (
            candidate['candidate_id'],
            final_score
        )
    )
```

```python
# Sort
ranked_candidates.sort(
    key=lambda x: x[1],
    reverse=True
)

ranked_candidates[:10]
```

```python
#Career History score
good_terms = ["retrieval", "ranking","recommendation","search","embedding",
              "vector","llm","fine-tuning","rag"]

bad_terms = ["speech","tts","asr","computer vision",
             "object detection","robotics"]

def career_score(candidate):

    text = ""

    for job in candidate['career_history']:
        text += " "
        text += job['title'].lower()
        text += " "
        text += job['description'].lower()

    score = 0

    for term in good_terms:
        if term in text:
            score += 1

    for term in bad_terms:
        if term in text:
            score -= 1

    score = max(score, 0)

    return min(score / len(good_terms), 1)
```

```python
# Consulting Companies penalties
service_companies = ["tcs","infosys","wipro","cognizant","capgemini","accenture"]

def consulting_penalty(candidate):

    companies = [
        job['company'].lower()
        for job in candidate['career_history']
    ]

    count = 0

    for company in companies:
        if any(
            service in company
            for service in service_companies
        ):
            count += 1

    if count == len(companies):
        return 0.3

    return 0
```

```python
#Improvised Final Score
improved_ranked_candidates = []

for idx in top_indices:

    candidate = filtered_candidates[idx]

    semantic = similarities[idx]

    skill = skill_score(candidate)

    exp = experience_score(candidate)

    redrob = redrob_score(candidate)

    career = career_score(candidate)

    penalty = consulting_penalty(candidate)

    final_score = (
        0.35 * semantic
        + 0.20 * skill
        + 0.15 * exp
        + 0.15 * redrob
        + 0.15 * career
        - penalty
    )

    improved_ranked_candidates.append(
        (
            candidate['candidate_id'],
            final_score
        )
    )
```

```python
# Sort
improved_ranked_candidates.sort(
    key=lambda x: x[1],
    reverse=True
)

improved_ranked_candidates[:20]
```
